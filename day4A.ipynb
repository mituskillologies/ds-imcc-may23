### Multiple Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### import the dataset

```python
df = pd.read_csv('mtcars.csv')
```

```python
df.shape
```

```python
df
```

```python
df.corrwith(df['mpg'])
```

```python
salary = pd.read_csv('Salary_Data.csv')
```

```python
salary.corr()
```

#### separate the input and output variables

```python
x = df[['disp','hp','wt']]
y = df['mpg']
```

```python
x
```

#### import the class

```python
from sklearn.linear_model import LinearRegression
```

#### create the object

```python
reg = LinearRegression()
```

#### train the model

```python
reg.fit(x, y)
```

#### predict on unseen data

```python
disp = 221
hp = 102
wt = 3.51
reg.predict([[disp, hp, wt]])
```

#### visualize

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.scatter(df['disp'], y, color='r')
plt.subplot(2,2,2)
plt.scatter(df['hp'], y, color='b')
plt.subplot(2,2,3)
plt.scatter(df['wt'], y, color='g')
```

```python
reg.score(x, y)
```

### Decision Tree Classification

#### import the dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
list(df.columns)
```

```python
df
```

#### separate the input and output variables

```python
x = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

#### explore the data

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], c=y)
```

#### import the class

```python
from sklearn.tree import DecisionTreeClassifier
```

#### create the object

```python
classifier = DecisionTreeClassifier(random_state=0)
```

#### train the model

```python
classifier.fit(x, y)
```

#### plot the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,15))
plot_tree(classifier, fontsize=7, feature_names=['age','sal'],
         class_names=['y','n'], filled=True, rounded=True);
```

#### Prediction on unseen data

```python
new = pd.DataFrame({
    'Age': [25,37,56,66],
    'EstimatedSalary': [34000,41000,26000,134000]
})
```

```python
new
```

```python
classifier.predict(new)
```

### Naive Bayes Classification

```python
from sklearn.naive_bayes import GaussianNB
```
