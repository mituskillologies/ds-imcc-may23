### Line Plot

```python
x = np.array([3,13,19,31,38,45,58,65,71,88])
y = np.array([11,16,24,37,41,58,56,78,81,95])
z = np.array([81,80,79,63,54,32,29,26,18,14])
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
# markers: .  o  x  v  >  ^  <  D  * 
plt.plot(x, y, color='red', marker='o', ls=':', mfc='green')
```

### Bar plot

```python
enrollment = [104,112,256,178,126,180]
courses = ['C','Java','Python','R','PHP','JavaScript']
```

```python
plt.xlabel('Courses')
plt.ylabel('Enrolled Students')
plt.grid()
plt.bar(courses, enrollment, color='brown')
```

```python
plt.ylabel('Courses')
plt.xlabel('Enrolled Students')
plt.grid()
plt.barh(courses, enrollment, color='brown')
```

### Pie chart

```python
plt.figure(figsize=(10, 10))
plt.pie(enrollment, labels=courses, autopct='%2.2f', shadow=True);
```

### Regression

```python
# Dataset: Salary_Data.csv, mtcars.csv, Social_Network_Ads.csv
#          Mall_Customers.csv     
# Location: https://mitu.co.in/dataset
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df.shape
```

```python
df
```

```python
plt.scatter(df['YearsExperience'], df['Salary'])
```

#### separate the input and output data

```python
x = df['YearsExperience'].values
y = df['Salary']
```

```python
x
```

```python
y;
```

```python
x = x.reshape(30,1)
```

```python
x
```

#### import the linear regression class

```python
from sklearn.linear_model import LinearRegression
```

#### create the object of linear regression

```python
regressor = LinearRegression()
```

#### train the algorithm

```python
regressor.fit(x, y)
```

#### predict on the unseen data

```python
regressor.predict([[15]])
```

```python
regressor.predict([[20]])
```

```python
regressor.score(x, y)
```
